```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import matplotlib.pyplot as plt

#load data
df_data = pd.read_csv('temp.csv', index_col=0, parse_dates=True)


#Train_test-split
df = df_data.head(52500) #create train df (df_t)with all values expect one year

Xtrain = df.index
ytrain = df['t_mean']


# Analyzing Time Series
plt.plot(Xtrain, ytrain)

# Trend
df['timestep'] = range(len(df)) #creating timesteps to get the trend out of the data with a LinearRegression

Xtrend = df[['timestep']]#just use the trainingdata

m = LinearRegression()
m.fit(Xtrend, ytrain)
print(m.coef_) #getting a trend of 3.15358284e-05 (the slope of the linear behaviour)
df['trend'] = m.predict(Xtrend) #create trend column in df_t

df[['t_mean', 'trend']].plot()#plot the trend

# Seasonality
df['month'] = df.index.month #create column with no. of month

df.groupby(df.index.month)['t_mean'].mean().plot() #plot mean Temp value of each month for the time series
print('more or less a normal Distribution with max at Jul. small tailing at beginningf of year.')

seasonal_dummies = pd.get_dummies(df['month'],prefix='month_') # create dummies for each month

df = df.merge(seasonal_dummies,left_index = True, right_index=True) #merge dummies df and df

Xseason = df.drop(['t_mean','trend','month'], axis=1) #choose X values for LinReg fit (all dummy columns)
m.fit(Xseason, ytrain)

df['trendANDseasonality'] = m.predict(Xseason) #create trend and seasonality column in df
df[['t_mean','trendANDseasonality']].plot()#plot the seasonality and trend
'min and max are not displayed by this way of getting the seasonality'

# Remainder
df['remainder'] = df['t_mean'] - df['trendANDseasonality'] #calculate the remainder by subtracting trend ans seasonality
df['remainder'].plot() # plot the remainder
df.head() 
```
