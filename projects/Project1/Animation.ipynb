```python
import pandas as pd
```

```python
fert = pd.read_csv('gapminder_total_fertility.csv', index_col=0)
```

```python
life = pd.read_excel('gapminder_lifeexpectancy.xlsx', index_col=0)
```

```python
print(fert.shape)
print(life.shape)
```

```python
fert.columns
```

```python
life.columns
```

```python
fert.columns = fert.columns.astype(int)
```

```python
fert.columns 
```

```python
fert.index
```

```python
fert.index.name = 'country'
```

```python
fert = fert.reset_index()
```

```python
fert = fert.melt(id_vars='country', var_name='year', value_name='fertility_rate')
```

```python
life.index
life.index.name = 'country'
life = life.reset_index()
life = life.melt(id_vars='country', var_name='year', value_name='life_expectancy')
```

```python
population = pd.read_excel('gapminder_population.xlsx', index_col=0)
```

```python
print(population.shape)
population.columns
```

```python
population
```

```python
population.index
population.index.name = 'country'
```

```python
population = population.reset_index()
```

```python
population
```

```python
population = population.melt(id_vars='country', var_name='year', value_name='total_population')
```

```python
population
```

```python
df = fert.merge(population, on = ['country', 'year'])
```

```python
df = df.merge(life, on = ['country', 'year'])
```

```python
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
sns.set_theme(style="darkgrid")
```

```python
df_subset = df.loc[df['country'].isin(['Spain', 'Canada', 'China'])]
sns.lineplot(x='year', y='life_expectancy', hue='country',data=df_subset)          
```

```python
sns.scatterplot(x='year', y='life_expectancy', hue='country',data=df_subset, alpha=0.6)
```

```python
for n in range(1960,2015+1):
    df_anm = df.loc[df['year'] == n]
    ax = sns.scatterplot(x='fertility_rate', y='life_expectancy', data=df_anm, alpha=0.7, 
                             size= df['total_population'],legend=False, palette="ch:r=-.5,l=.75", sizes=(37, 370))
    ax.set(xlabel=('life_expectancy'), ylabel=('fertility_rate'))
    plt.axis((0,11,0,107))
    plt.savefig('lifeexp_{0}.png'.format(n))
```
